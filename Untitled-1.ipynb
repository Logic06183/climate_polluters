```python
# Load the CSV data into a pandas DataFrame
data = pd.read_csv('emissions_high_granularity.csv')

# Display the first few rows of the DataFrame
print(data.head())
```

```python
# Group the data by parent entity and sum the total emissions
global_polluters_grouped = data.groupby('parent_entity')['total_emissions_MtCO2e'].sum().reset_index()

# Sort by total emissions in descending order
global_polluters_sorted = global_polluters_grouped.sort_values(by='total_emissions_MtCO2e', ascending=False)

# Display the top 10 global polluters and their contribution
top_10_global_polluters = global_polluters_sorted.head(10)
top_10_global_polluters
```

```python
# Display the top 100 global polluters and their contribution
top_100_global_polluters = global_polluters_sorted.head(100)
top_100_global_polluters
```

```python
import matplotlib.pyplot as plt

# Plot the top 100 global polluters
plt.figure(figsize=(12, 8))
plt.barh(top_100_global_polluters['parent_entity'][:20], top_100_global_polluters['total_emissions_MtCO2e'][:20])
plt.xlabel('Total Emissions (MtCO2e)')
plt.ylabel('Parent Entity')
plt.title('Top 20 Global Polluters Based on Total Emissions')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest emitter at the top
plt.show()
```

```python
# Filter the data for Sasol
sasol_data = data[data['parent_entity'].str.contains('Sasol', case=False)]

# Group by year and sum the total emissions
sasol_emissions_by_year = sasol_data.groupby('year')['total_emissions_MtCO2e'].sum().reset_index()

# Display Sasol's emissions data
sasol_emissions_by_year
```

```python
# Plot Sasol's emissions over time
plt.figure(figsize=(12, 8))
plt.plot(sasol_emissions_by_year['year'], sasol_emissions_by_year['total_emissions_MtCO2e'], marker='o')
plt.xlabel('Year')
plt.ylabel('Total Emissions (MtCO2e)')
plt.title('Sasol Total Emissions Over Time')
plt.grid(True)
plt.show()
```

```python
# Display detailed emissions data for Sasol
sasol_detailed_data = sasol_data[['year', 'product_emissions_MtCO2', 'flaring_emissions_MtCO2', 'venting_emissions_MtCO2', 
                                  'own_fuel_use_emissions_MtCO2', 'fugitive_methane_emissions_MtCO2e', 'total_operational_emissions_MtCO2e', 
                                  'total_emissions_MtCO2e']]
sasol_detailed_data.tail(10)  # Display the last 10 years for a more detailed view
```

Here's a breakdown of Sasol's emissions for the last 10 years, with detailed categories:

Product Emissions (MtCO2): Emissions from the use of sold products.
Flaring Emissions (MtCO2): Emissions from the flaring of gases.
Venting Emissions (MtCO2): Emissions from the venting of gases.
Own Fuel Use Emissions (MtCO2): Emissions from the combustion of fuel for energy.
Fugitive Methane Emissions (MtCO2e): Emissions from leaks or releases of methane gas.
Total Operational Emissions (MtCO2e): Sum of all operational emissions.
Total Emissions (MtCO2e): Overall sum of emissions, including product emissions and operational emissions.

```python
# Aggregate the emissions data for Sasol over the last 10 years
sasol_aggregated_data = sasol_detailed_data.groupby('year').sum().reset_index()

# Plot the breakdown of Sasol's emissions
plt.figure(figsize=(14, 10))
plt.stackplot(sasol_aggregated_data['year'], 
              sasol_aggregated_data['product_emissions_MtCO2'],
              sasol_aggregated_data['flaring_emissions_MtCO2'],
              sasol_aggregated_data['venting_emissions_MtCO2'],
              sasol_aggregated_data['own_fuel_use_emissions_MtCO2'],
              sasol_aggregated_data['fugitive_methane_emissions_MtCO2e'],
              labels=['Product Emissions', 'Flaring Emissions', 'Venting Emissions', 'Own Fuel Use Emissions', 'Fugitive Methane Emissions'],
              colors=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])
plt.xlabel('Year')
plt.ylabel('Emissions (MtCO2e)')
plt.title('Breakdown of Sasol Emissions Over Time')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'data' is the DataFrame with all the emissions data.
# Here is how you would filter and plot the data for the specified companies.

# List of South African or South Africa-related companies
sa_companies = [
    'Sasol',
    'Anglo American',
    'Exxaro Resources Ltd',
    'Seriti Resources'
]

# Dictionary to hold emissions data for each company
emissions_data = {}

# Loop through each company to filter data and calculate total emissions per year
for company in sa_companies:
    company_data = data[data['parent_entity'].str.contains(company, case=False)]
    emissions_data[company] = company_data.groupby('year')['total_emissions_MtCO2e'].sum()

# Convert to DataFrame for easier plotting
emissions_df = pd.DataFrame(emissions_data)

# Plotting
plt.figure(figsize=(14, 7))

for company in sa_companies:
    plt.plot(emissions_df.index, emissions_df[company], label=company, marker='o')

plt.title('CO2 Emissions Over Time for Selected South African Companies')
plt.xlabel('Year')
plt.ylabel('Total Emissions (MtCO2e)')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Assuming 'data' is your DataFrame containing the emissions data
# Let's create a dictionary to hold emissions data for each commodity by company

# Define the South African or related companies again
sa_companies = [
    'Sasol',
    'Anglo American',
    'Exxaro Resources Ltd',
    'Seriti Resources'
]

# Dictionary to hold emissions data for each commodity by company
commodity_emissions_data = {}

# Loop through each company
for company in sa_companies:
    company_data = data[data['parent_entity'].str.contains(company, case=False)]
    # Now, group by year and commodity and sum the total emissions
    commodity_emissions = company_data.groupby(['year', 'commodity'])['total_emissions_MtCO2e'].sum().unstack().fillna(0)
    commodity_emissions_data[company] = commodity_emissions

# Now you have emissions data split by commodity for each company
# Let's plot this data

plt.figure(figsize=(14, 7))

# We will plot data for each commodity of each company in a loop
for company, emissions_df in commodity_emissions_data.items():
    for commodity in emissions_df.columns:
        plt.plot(emissions_df.index, emissions_df[commodity], label=f"{company} - {commodity}", marker='o')

plt.title('CO2 Emissions Over Time by Commodity for South African Companies')
plt.xlabel('Year')
plt.ylabel('Total Emissions (MtCO2e)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Assuming 'data' is your DataFrame containing the emissions data.

# Define South African companies again
sa_companies = [
    'Sasol',
    'Anglo American',
    'Exxaro Resources Ltd',
    'Glencore',
    'BHP',
    'Seriti Resources'
]

# Dictionary to hold a DataFrame for each commodity
commodity_rankings = {}

# Get the unique list of commodities in the dataset
commodities = data['commodity'].unique()

# For each commodity, calculate the total emissions and rank the companies
for commodity in commodities:
    # Filter the dataset for the current commodity
    commodity_data = data[data['commodity'] == commodity]
    
    # Group by parent entity and sum the emissions, then sort and reset index
    commodity_ranking = (commodity_data.groupby('parent_entity')['total_emissions_MtCO2e']
                         .sum()
                         .sort_values(ascending=False)
                         .reset_index())
    
    # Find the rank of each South African company for this commodity
    for company in sa_companies:
        company_rank = commodity_ranking[commodity_ranking['parent_entity'].str.contains(company, case=False, na=False)]
        
        # If the company is found in the rankings, add it to the dictionary
        if not company_rank.empty:
            # Create an entry in the dictionary if not present
            if commodity not in commodity_rankings:
                commodity_rankings[commodity] = []
            
            # Append the rank information
            commodity_rankings[commodity].append(company_rank)

# commodity_rankings now contains the ranks of SA companies for each commodity
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the commodity you are interested in
commodity_of_interest = 'Oil & NGL'  # Replace with the actual commodity of interest

# Filter the data for the chosen commodity
commodity_data = data[data['commodity'] == commodity_of_interest]

# Group by parent entity and sum the emissions, then sort and reset index
commodity_ranking = (commodity_data.groupby('parent_entity')['total_emissions_MtCO2e']
                     .sum()
                     .sort_values(ascending=False)
                     .reset_index())

# Find the positions of South African companies
positions = []
for company in sa_companies:
    company_position = commodity_ranking[commodity_ranking['parent_entity'].str.contains(company, case=False, na=False)]
    if not company_position.empty:
        # Extract the position
        position = commodity_ranking.index[commodity_ranking['parent_entity'] == company_position.iloc[0]['parent_entity']].tolist()[0]
        positions.append((company, position))

# Sort positions by rank
positions.sort(key=lambda x: x[1])

# Set the style
sns.set(style="whitegrid")

# Create a bar chart
plt.figure(figsize=(14, 10))

# Plot all companies
sns.barplot(x='total_emissions_MtCO2e', y='parent_entity', data=commodity_ranking, color='lightgray')

# Highlight South African companies
for company, position in positions:
    plt.barh(commodity_ranking.iloc[position]['parent_entity'], commodity_ranking.iloc[position]['total_emissions_MtCO2e'], color='orange')

plt.title(f'Global Ranking of Total Emissions for {commodity_of_interest}', fontsize=16)
plt.xlabel('Total Emissions (MtCO2e)', fontsize=14)
plt.ylabel('Parent Entity', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import panel as pn
import matplotlib.pyplot as plt
import numpy as np
pn.extension()


# Define a simple plotting function
def plot_sine_wave(frequency=1.0):
    plt.figure(figsize=(10, 6))
    t = np.linspace(0, 10, 500)
    plt.plot(t, np.sin(2 * np.pi * frequency * t))
    plt.title(f'Sine Wave with Frequency {frequency} Hz')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.close()  # Prevent Matplotlib from showing the figure separately
    return pn.pane.Matplotlib(plt.gcf())

# Create a Panel interactive object
frequency_slider = pn.widgets.FloatSlider(name='Frequency', start=0.1, end=5, step=0.1, value=1)
interactive_sine_wave = pn.interact(plot_sine_wave, frequency=frequency_slider)

# Display the interactive dashboard
interactive_sine_wave

```
